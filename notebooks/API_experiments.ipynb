```python
import os
import deepl
from apify_client import ApifyClient
from dotenv import load_dotenv


load_dotenv()

deeplapi_key = os.getenv("DEEPLAPI_KEY")
```

```python
# Method for translating review text with Deepl API

def translate_text(reviews_list: list) -> list:
    deepl_client = deepl.DeepLClient(deeplapi_key)

    results = deepl_client.translate_text(reviews_list, target_lang="EN-US")
    translated_reviews = [result.text for result in results]

    return translated_reviews
```

#### Test Apify Api

```python
import sys
import os

notebook_dir = os.path.abspath(os.getcwd())
project_root = os.path.dirname(notebook_dir)

sys.path.append(project_root)

import json
from config.paths import DATA_DIR
```

```python
hotel_path = DATA_DIR / 'hotels.json'
TOKEN = os.getenv('APIFY_TOKEN')

def extract_reviews():

    with open(hotel_path, 'r', encoding='utf-8') as f:
        hotels = json.load(f) 

    apify_client = ApifyClient(TOKEN)
    actor_client = apify_client.actor('maxcopell/tripadvisor-reviews')

    # Define input for the actor
    run_input = {
        "startUrls": [{"url": hotel["url"]} for hotel in hotels],
        "maxItemsPerQuery": 1000000,
        "scrapeReviewerInfo": True
    }

    # Start the actor
    call_result = actor_client.call(run_input=run_input)

    if call_result is None:
        print('Actor run failed')
        return
    
    # Fetch results from the Actor run's default dataset 
    dataset_client = apify_client.dataset(call_result['defaultDatasetId'])
    list_items_result = dataset_client.list_items()
    print(list_items_result.items)
```
